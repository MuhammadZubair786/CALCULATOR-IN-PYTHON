```python
def add(a,b):
   
    c=a+b
    return c

def sub(a,b):
    
    c=a-b
    return c

def multi(a,b):
    
    c=a*b
    return c

def divis(a,b):

    c=a/b
    return c

def floordiv(a,b):

    c=a//b
    return c

def power(a,b):

    c=a**b
    return c

a='y'
while (a=='y' or a=='Y'):
    sel=int(input("1 FOR ADDITION\n2 FOR SUBTRACTION\n3 FOR MULTIPLICATION\n4 FOR DIVISION\n5 FOR FLOOR DIVISION\n6 FOR POWER\n\n SELECT OPERATION :"))
    
    if(sel==1):
        print("\nYou Select Addition Operation  ")
        a=int(input("Enter Value Of A :"))
        b=int(input("Enter Value Of B: "))
        result=add(a,b)
        print("Result Of ",a,"+",b,"is :",result)
        print("***************************")
        
    elif(sel==2):
        print("\nYou Select Subtraction Operation  ")
        a=int(input("Enter Value Of A :"))
        b=int(input("Enter Value Of B: "))
        result=sub(a,b)
        print("Result Of ",a,"-",b,"is :",result)
        print("***************************")
    
    elif(sel==3):
        print("\nYou Select Multiplication Operation  ")
        a=int(input("Enter Value Of A :"))
        b=int(input("Enter Value Of B: "))
        result=multi(a,b)
        print("Result Of ",a,"*",b,"is :",result)
        print("***************************")
        
    elif(sel==4):
        print("\nYou Select Division Operation  ")
        a=int(input("Enter Value Of A :"))
        b=int(input("Enter Value Of B: "))
        result=divis(a,b)
        print("Result Of ",a,"/",b,"is :",result)
        print("***************************")
        
    elif(sel==5):
        print("\nYou Select Floor Division Operation  ")
        a=int(input("Enter Value Of A :"))
        b=int(input("Enter Value Of B: "))
        result=floordiv(a,b)
        print("Result Of ",a,"//",b,"is :",result)
        print("***************************")
        
    elif(sel==6):
        print("\nYou Select Power Operation  ")
        a=int(input("Enter Base  Of A :"))
        b=int(input("Enter Power  Of B: "))
        result=power(a,b)
        print("Result Of",a,"^",b,"is :",result)
        print("***************************")
        
    else:
        print("\nYou Select Incorrect Operation  ")
        print("***************************")
       
    a=input("If Continue Then Press Y or y otherwise Enter Other Key :")
    
```
